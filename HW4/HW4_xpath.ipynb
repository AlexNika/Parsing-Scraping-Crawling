#### Практическое задание ####

Написать приложение, которое собирает основные новости с сайтов mail.ru, lenta.ru.
Для парсинга использовать xpath. Структура данных должна содержать:
* название источника,
* наименование новости,
* ссылку на новость,
* дата публикации

```python
from lxml import html
import requests
from pprint import pprint
from datetime import datetime
import pandas as pd
```

```python
months = {'января':'01', 'февраля':'02', 'марта':'03', 'апреля':'04', 'мая':'05', 'июня':'06',
          'июля':'07', 'августа':'08', 'сентября':'09', 'октября':'10', 'ноября':'11', 'декабря':'12'}
```

```python
def lrnews_time_convert(news_date):
    news_date = news_date.lstrip().split(',')
    news_date[1] = news_date[1].lstrip().split()
    news_date[1][1] = months.get(news_date[1][1])
    news_date[1] = '-'.join(news_date[1])
    news_date.reverse()
    return ' '.join(news_date)
```

```python
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) \
Chrome/77.0.3865.120 Safari/537.36'}
mr_url='https://mail.ru'
lr_url='https://lenta.ru'
```

```python
mr_request = requests.get(mr_url,headers=header)
lr_request = requests.get(lr_url,headers=header)
```

```python
mr_root = html.fromstring(mr_request.text)
lr_root = html.fromstring(lr_request.text)
```

```python
mr_news_hrefs = mr_root.xpath('//div[@class="news-item o-media news-item_media news-item_main"]/a/@href | \
//div[@class="news-item__inner"]/a[contains(@href, "https://")]/@href')
mr_news_text = mr_root.xpath('//div[@class="news-item o-media news-item_media news-item_main"]\
//h3[@class="news-item__title i-link-deco"]/text() | //div[@class="news-item__inner"]/ \
a[contains(@href, "https://")]/text()')
```

```python
mr_news_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
mailru_news = {'Source':mr_url[8:], 'News Title':mr_news_text, 'News URL':mr_news_hrefs, 'News time': mr_news_time}
mr_dataframe = pd.DataFrame(data=mailru_news)
mr_dataframe.index = mr_dataframe['News time']
mr_dataframe.drop('News time', axis=1, inplace=True)
mr_dataframe.head(16)
```

```python
lr_news_hrefs = lr_root.xpath('//section[@class="row b-top7-for-main js-top-seven"]//div[@class="first-item"]\
/h2/a/@href | //section[@class="row b-top7-for-main js-top-seven"]//div[@class="item"]/a/@href')
lr_news_text = lr_root.xpath('//section[@class="row b-top7-for-main js-top-seven"]//div[@class="first-item"]\
/h2/a/text() | //section[@class="row b-top7-for-main js-top-seven"]//div[@class="item"]/a/text()')
lr_news_time = lr_root.xpath('//section[@class="row b-top7-for-main js-top-seven"]//div[@class="first-item"]\
/h2/a/time/@datetime | //section[@class="row b-top7-for-main js-top-seven"]\
//div[@class="item"]/a/time/@datetime')
```

```python
for i in range(len(lr_news_hrefs)):
    if lr_news_hrefs[i][0] == '/':
        lr_news_hrefs[i] = lr_url + lr_news_hrefs[i]
for i in range(len(lr_news_time)):
    lr_news_time[i] = lrnews_time_convert(lr_news_time[i])
    lr_news_time[i] = datetime.strptime(lr_news_time[i], '%d-%m-%Y %H:%M')
lentaru_news = {'Source':lr_url[8:], 'News Title':lr_news_text, 'News URL':lr_news_hrefs, 'News time':lr_news_time}
lr_dataframe = pd.DataFrame(data=lentaru_news)
lr_dataframe.index = lr_dataframe['News time']
lr_dataframe.drop('News time', axis=1, inplace=True)
lr_dataframe.head(10)
```

```python
df_news = mr_dataframe.append(lr_dataframe)
df_news
```
