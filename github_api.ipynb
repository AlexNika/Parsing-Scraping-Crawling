```python
import json
import requests
from pprint import pprint
```

```python
username = 'AlexNika'
access_token='6b021c5cc84148609938b25214ed496623184234'
main_url = 'https://api.github.com'
repos_url = '{0}{1}{2}{3}'.format(main_url, '/users/', username, '/repos')
```

```python
request = requests.get(repos_url)
```

```python
pprint(request.json())
```

```python
data = request.json()
for i in range(len(data)):
    print('Project Number:', i+1)
    print('Project Name:', data[i]['name'])
    print('Project URL:', data[i]['html_url'],'\n')
with open('my_public_repo.json','w') as file:
    file.write(json.dumps(data, indent=4))
file.close()
```
